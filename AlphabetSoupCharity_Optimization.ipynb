```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
application_df = application_df.drop(columns = ["EIN", "NAME", "STATUS"])
application_df.head()
```

```python
application_df.nunique()
```

```python
app_type = application_df["APPLICATION_TYPE"].value_counts()
app_type
```

```python
application_df = application_df.drop(columns = ["ASK_AMT"])
application_df.head()
```

```python
application_df.nunique()
```

```python
app_type = application_df["APPLICATION_TYPE"].value_counts()
app_type
```

```python
application_types_to_replace = list(app_type.iloc[8:].index)
cutoff_value = 550

for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

application_df['APPLICATION_TYPE'].value_counts()
```

```python
class_value = application_df["CLASSIFICATION"].value_counts()
class_value.head()
```

```python
class_filtered = class_value.loc[class_value > 2]
class_filtered
```

```python
classifications_to_replace = list(class_value.iloc[5:].index)
cutoff_value = 1000

for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

application_df['CLASSIFICATION'].value_counts()
```

```python
numeric = pd.get_dummies(application_df)
numeric.head()
```

```python
y = numeric["IS_SUCCESSFUL"].values
X = numeric.drop(columns = "IS_SUCCESSFUL").values

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
input_features = len(X_train[0])
hidden_nodes_1 = 8
hidden_nodes_2 = 5

nn = tf.keras.models.Sequential()

nn.add(tf.keras.layers.Dense(units = hidden_nodes_1, input_dim = input_features, activation = "relu"))

nn.add(tf.keras.layers.Dense(units = hidden_nodes_2,  activation = "relu"))

nn.add(tf.keras.layers.Dense(units = 1, activation = "sigmoid"))

nn.summary()
```

```python
nn.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

```python
train_model = nn.fit(X_train_scaled, y_train, epochs = 40)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
